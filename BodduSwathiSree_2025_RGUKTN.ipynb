#### Import necessary libraries

```python
import turtle
import math
```

#### Utility functions

```python
#Draws a filled or outlined circle(ring)
def draw_ring(t,radius,color,thickness=0):
    t.penup()
    t.goto(0,-radius)
    t.pendown()
    if thickness==0:
        t.color(color)
        t.begin_fill()
        t.circle(radius)
        t.end_fill()
    else:
        t.pensize(thickness)
        t.color(color)
        t.circle(radius)
        t.pensize(1)
    t.penup()
    t.home()
    t.pendown()

#Draws green triangular petals around circle
def outer_green_petals(t,radius,count,size,color):
    angle_step=360/count
    t.color(color)
    for i in range(count):
        angle_rad=math.radians(i*angle_step)
        x=math.cos(angle_rad)*radius
        y=math.sin(angle_rad)*radius
        t.penup()
        t.goto(x,y)
        t.setheading(i*angle_step-90)
        t.pendown()
        t.begin_fill()
        for j in range(3):
            t.forward(size)
            t.left(120)
        t.end_fill()
        t.penup()
        t.home()
        t.pendown()

#Draws a filled continous square
def square(t,size,color):
    t.color(color)
    t.begin_fill()
    for i in range(4):
        t.forward(size)
        t.right(90)
    t.end_fill()

# Draws small chocolate petals around the outer red ring
def chocolate_petals_around_outer_red(t,radius,count=48):
    angle_step=360/count
    for i in range(count):
        t.penup()
        t.goto(0,0)
        t.setheading(i*angle_step)
        t.forward(radius)
        t.pendown()
        t.color("chocolate")
        t.begin_fill()
        t.forward(15)
        t.right(90)
        t.forward(8)
        t.right(90)
        t.forward(15)
        t.end_fill()
        t.penup()
        t.home()

#Draws pentagonal orange petals outside a ring
def outer_red_pentagonal_petals(t,radius,count,size,color,outward_offset=1):
    angle_step=360/count
    t.color(color)
    for i in range(count):
        angle_rad=math.radians(i*angle_step)
        x=math.cos(angle_rad)*(radius+outward_offset)
        y=math.sin(angle_rad)*(radius+outward_offset)
        t.penup()
        t.goto(x,y)
        t.setheading(i*angle_step-90)
        t.pendown()
        t.begin_fill()
        for j in range(5):
            t.forward(size)
            t.left(72)
        t.end_fill()
        t.penup()
        t.home()
        t.pendown()

#Draws a palegoldenrod ring between green and brown petals
def draw_palegoldenrod_ring(t,green_radius,petal_size,brown_outward_offset,step_up=0,color="palegoldenrod"):
    tip_offset=petal_size/(2*math.sin(math.pi/5))
    inner_radius=green_radius+step_up
    outer_radius=green_radius+tip_offset+brown_outward_offset+step_up
    thickness=outer_radius-inner_radius
    ring_radius=inner_radius+thickness/2
    t.penup()
    t.goto(0,-ring_radius)
    t.pendown()
    t.pensize(thickness)
    t.pencolor(color)
    t.circle(ring_radius)
    t.pensize(1)
    t.penup()
    t.home()
    t.pendown()
    return outer_radius

#Draws a brown ring outside the palegoldenrod ring
def draw_outer_brown_ring(t,radius,petal_size,color="saddlebrown",thickness=10,outward_offset=23):
    tip_offset=petal_size/(2*math.sin(math.pi/5))
    ring_radius=radius+tip_offset+outward_offset
    t.penup()
    t.goto(0,-ring_radius)
    t.pendown()
    t.pensize(thickness)
    t.pencolor(color)
    t.circle(ring_radius)
    t.pensize(1)
    t.penup()
    t.home()
    t.pendown()
    return ring_radius+thickness/2

#Draws an outer orange ring
def draw_outer_orange_ring(t,brown_outer_radius,thickness=15):
    t.penup()
    t.goto(0,-brown_outer_radius)
    t.pendown()
    t.pensize(thickness)
    t.pencolor("orange")
    t.circle(brown_outer_radius)
    t.pensize(1)
    t.penup()
    t.home()
    t.pendown()

#Draws maroon petals inside the orange ring
def maroon_petals_in_orange(t,radius,count=24):
    angle_step=360/count
    for i in range(count):
        t.penup()
        t.goto(0,0)
        t.setheading(i*angle_step)
        t.forward(radius)
        t.pendown()
        t.color("maroon")
        t.begin_fill()
        t.forward(15)
        t.right(90)
        t.forward(8)
        t.right(90)
        t.forward(15)
        t.end_fill()
        t.penup()
        t.home()

#Draws small green triangle petals
def darkgreen_triangle_petals(t,ring_radius,count,size):
    angle_step=360/count
    t.color("darkgreen")
    for i in range(count):
        angle_rad=math.radians(i*angle_step)
        x=math.cos(angle_rad)*ring_radius
        y=math.sin(angle_rad)*ring_radius
        t.penup()
        t.goto(x,y)
        t.setheading(i*angle_step-90)
        t.pendown()
        t.begin_fill()
        for j in range(3):
            t.forward(size)
            t.left(120)
        t.end_fill()
        t.penup()
        t.home()
        t.pendown()

#Draws concentric flower petals alternating between two colors
def flower_petals(t,radius,step,color1,color2):
    angle=0
    i=0
    while angle<360:
        t.color(color1 if i%2==0 else color2)
        t.begin_fill()
        t.circle(radius,60)
        t.left(120)
        t.circle(radius,60)
        t.end_fill()
        angle+=step
        if angle<=360:
            t.right(step)
        i+=1
    t.penup()
    t.home()
    t.pendown()
```

#### 'main' function

```python
def main():
    red_radius=135
    triangle_size=25
    border_thickness=10
    outer_petals_offset=18
    petal_size=40
    brown_outward_offset=23
    extra_offset=13
    #Precalculate radii based on geometry
    outer_petals_radius=red_radius+triangle_size*math.sqrt(3)/2+border_thickness+outer_petals_offset
    outer_brown_radius=outer_petals_radius+petal_size/(2*math.sin(math.pi/5))+brown_outward_offset+extra_offset+10
    #Setup screen and turtle
    s=turtle.Screen()
    s.setup(width=700,height=700)
    s.title("BodduSwathiSree_2025_RGUKTN")
    s.bgcolor("white")
    t=turtle.Turtle()
    t.color("brown")
    t.speed(0)
    #Draws base rings and petals
    draw_ring(t,red_radius,"red")
    draw_ring(t,red_radius+triangle_size*math.sqrt(3)/2/2,"palegoldenrod",thickness=triangle_size*math.sqrt(3)/2)
    outer_green_petals(t,red_radius,24,triangle_size,"darkgreen")
    #Draws redouter circle
    draw_ring(t,red_radius+triangle_size*math.sqrt(3)/2+(border_thickness+5)/2,"red",thickness=border_thickness+5)
    draw_ring(t,red_radius+triangle_size*math.sqrt(3)/2+border_thickness+19/2+1,"palegoldenrod",thickness=19)
    chocolate_petals_around_outer_red(t,red_radius+triangle_size*math.sqrt(3)/2+border_thickness)
    t.penup()
    t.goto(0,-outer_petals_radius)
    t.pendown()
    t.pensize(10)
    t.pencolor("darkgreen")
    t.circle(outer_petals_radius)
    t.pensize(1)
    t.penup()
    t.home()
    t.pendown()
    #Draw outer palegoldenrod ring
    green_ring_radius=outer_petals_radius
    palegold_outer=draw_palegoldenrod_ring(t,green_ring_radius,petal_size,brown_outward_offset+extra_offset,step_up=5,color="palegoldenrod")
   
    outer_red_pentagonal_petals(t,green_ring_radius,18,petal_size,"red",outward_offset=4)
    brown_offset_adjust=0.5
    brown_outer_radius=draw_outer_brown_ring(t,green_ring_radius,petal_size,color="saddlebrown",thickness=20,
                                             outward_offset=brown_outward_offset+extra_offset+brown_offset_adjust)
   
    draw_outer_orange_ring(t,brown_outer_radius+5,thickness=15)

    orange_inner_edge=brown_outer_radius+5-15/2
    offset=3
    maroon_radius=orange_inner_edge-offset
    maroon_petals_in_orange(t,maroon_radius,count=24)
    # Draw base rings and squares
    orange_outer_radius=brown_outer_radius+5+15/2
    palegold_thickness=20
    offset=-1
    palegold_radius=orange_outer_radius+offset+palegold_thickness/2
    draw_ring(t,palegold_radius,"palegoldenrod",thickness=palegold_thickness)

    palegold_inner_edge=palegold_radius-palegold_thickness/2
    darkgreen_triangle_count=24
    darkgreen_triangle_size=25
    darkgreen_triangle_petals(t,palegold_inner_edge,darkgreen_triangle_count,darkgreen_triangle_size)
    #Draws maroon and brown rings
    maroon_thickness=8
    maroon_offset=-1
    maroon_radius=palegold_radius+palegold_thickness/2+maroon_offset+maroon_thickness/2
    draw_ring(t,maroon_radius,"maroon",thickness=maroon_thickness)

    brown_thickness=8
    brown_offset=-0.8
    brown_radius=maroon_radius+maroon_thickness/2+brown_offset+brown_thickness/2
    draw_ring(t,brown_radius,"saddlebrown",thickness=brown_thickness)
    #Draws base rings and squares
    draw_ring(t,red_radius-triangle_size*math.sqrt(3)/2,"saddlebrown",thickness=15)
    draw_ring(t,127,"palegoldenrod")
    draw_ring(t,120,"darkorange")

    for i in range(12):
        square(t,90,"gold")
        t.left(30)
    draw_ring(t,90,"orange")
    draw_ring(t,70,"palegoldenrod")
    # Draw flower petals at centre
    t.hideturtle()
    t.color("maroon")
    flower_petals(t,60,30,"maroon","chocolate")
    flower_petals(t,35,30,"bisque","brown")
    turtle.done()
```

```python
main()
```
